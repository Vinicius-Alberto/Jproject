<H2> Empresas que mais venderamm Globalmente</H2>

```python
df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 6))
df['Year'].value_counts().sort_index().plot(kind='line', marker='o', color='blue')
plt.xlabel("Ano")
plt.ylabel("Quantidade de Jogos LanÃ§ados")
plt.title("NÃºmero de Jogos LanÃ§ados por Ano")
plt.show()
```

```python
# Somando as vendas por regiÃ£o
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Criando o grÃ¡fico de anel
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', 
                                  startangle=140, colors=['red', 'blue', 'green', 'orange'], 
                                  wedgeprops=dict(width=0.4))  # Define a largura do anel

# Adicionando um cÃ­rculo branco no centro
plt.gca().add_artist(plt.Circle((0, 0), 0.3, color='white'))

# Ajustando a posiÃ§Ã£o dos nÃºmeros para ficarem dentro do anel
for autotext in autotexts:
    autotext.set_color('white')  # Define a cor do texto
    autotext.set_fontsize(12)     # Ajusta o tamanho da fonte
    autotext.set_weight("bold")   # Deixa o texto em negrito
    autotext.set_bbox(dict(facecolor='black', edgecolor='none', boxstyle='round,pad=0.2'))  # Adiciona um fundo escuro

plt.title("DistribuiÃ§Ã£o de Vendas por RegiÃ£o")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='NA_Sales', y='JP_Sales', hue='Publisher', legend=False)
plt.title('RelaÃ§Ã£o entre Vendas na AmÃ©rica do Norte e JapÃ£o')
plt.xlabel('NA Sales (milhÃµes)')
plt.ylabel('JP Sales (milhÃµes)')
plt.show()
```

<H2>DistribuiÃ§Ã£o de Vendas por publisher</H2>

```python
vendas_por_publisher = df.groupby("Publisher")["Global_Sales"].sum().reset_index()
vendas_por_publisher = vendas_por_publisher.sort_values(by="Global_Sales", ascending=False)

top5 = df[df['Publisher'].isin(vendas_por_publisher.head(5)['Publisher'])]

plt.figure(figsize=(10, 6))
sns.violinplot(data=top5, x='Publisher', y='Global_Sales', palette='pastel')
plt.title('DistribuiÃ§Ã£o das Vendas Globais por Publisher (Top 5)')
plt.ylabel('Global Sales (milhÃµes)')
plt.xticks(rotation=45)
plt.show()
```

```python
# GrÃ¡fico de barras horizontal com os top 10 publishers
plt.figure(figsize=(12, 5))
ax = sns.barplot(x="Global_Sales", y="Publisher", data=vendas_por_publisher.head(10), palette="viridis")

# Adiciona os valores DENTRO das barras
for bar in ax.patches:
    width = bar.get_width()
    ax.text(width - 20,  # PosiÃ§Ã£o X um pouco antes do fim da barra
            bar.get_y() + bar.get_height() / 2,  # PosiÃ§Ã£o Y no centro da barra
            f'{width:.1f}',  # Texto com 1 casa decimal
            ha='right', va='center', fontsize=10, color='white')  # Alinhamento Ã  direita e cor branca

plt.xlabel("Vendas Globais (MilhÃµes de Unidades)")
plt.ylabel("Publisher")
plt.title("Top 10 Publishers por Vendas Globais")
plt.tight_layout()
plt.show()
```
